### Task_4:

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5, 6.5, 10000)

plt.plot(x, x**2 - 1)
plt.plot(x, 1 - (1 - np.exp(x)) / x)
```

```python
def eq(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x * (1 - y) - 1)
```

```python
x1, y1 = fsolve(eq, (4, 19))
x1, y1
```

```python
X = []
for i in np.linspace(-20, 20, 1000):
    x1, y1 = fsolve(eq, (i, 10))
    X.append(round(x1, 5))
    

X = list(set(X))
Y = [x**2 + 1 for x in X]

X, Y
```
