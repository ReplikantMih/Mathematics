### Task_1.2:

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def length_calc(vector):
    sum = 0
    for coord in vector:
        sum += coord**2
    return math.sqrt(sum)

length_calc((1, 1, 7, 17, 121))
```

### Task_3:

3.1. (круг):

```python
x = np.linspace(-2, 2, 1000)

plt.figure(figsize=(5,5))
plt.plot(x, np.sqrt(4 - x**2), c='green')
plt.plot(x, -np.sqrt(4 - x**2), c='green')
```

3.2. (элипс):

```python
x = np.linspace(-2, 2, 10000)


plt.figure(figsize=(5,5))
plt.axis('equal')
plt.plot(x, np.sqrt(1**2 - x**2 * 1**2 / 2**2), c='green')
plt.plot(x, -np.sqrt(1**2 - x**2 * 1**2 / 2**2), c='green')
```

3.3. (гипербола):

```python
x = np.linspace(-10, 10, 10000)

plt.figure(figsize=(5,5))
plt.axis('equal')
plt.plot(x, np.sqrt(x**2 * 1**2 / 2**2 - 1**2), c='green')
plt.plot(x, -np.sqrt(x**2 * 1**2 / 2**2 - 1**2), c='green')
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)

fig = figure()
ax = Axes3D(fig)

ax.plot_wireframe(x, y, 2*x + 3*y + 1)
ax.plot_wireframe(x, y, 2*(x-5) + 3*y + 1)
```
